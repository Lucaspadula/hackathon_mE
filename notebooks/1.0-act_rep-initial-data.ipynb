```python
#Detectar si tiene valores NULL o Vacios
def detectar_null_vacio(df):
    valores_vacion_null = {}
    for column in df.columns:
        nulls = df[column].isnull().sum()
        empties = (df[column]== '').sum()
        total = nulls + empties
        if total > 0:
            valores_vacion_null[column] = total
            
    return valores_vacion_null
    

resultado = detectar_null_vacio(df)

if resultado:
    for campo,cantidad in  resultado.items():
        print(f'El campo {campo} tiene {cantidad} valores NULL o vacios')
else:
    print('No hay campos NULL o vacios ')
```

```python
for year in years:
    plt.figure(figsize=(15,5))
    
    # Filtrar datos por aÃ±o
    df_filtered = df[df['YEAR'] == year]
    
    # Agrupar y sumar las cantidades
    grouped_data = df_filtered.groupby('ID_REPARTIDOR').agg({
        'CANT_FACT_TOTALES': 'sum',
        'CANT_FACT_ENTREGA_TOTAL': 'sum'
    }).reset_index()
    
    # Crear grÃ¡fico
    ax = sns.barplot(
        x='ID_REPARTIDOR',
        y='CANT_FACT_TOTALES',
        data=grouped_data,
        color='black',
        label='Facturas Totales'
    )
    
    # Agregar barras para la cantidad de facturas entregadas
    ax2 = sns.barplot(
        x='ID_REPARTIDOR',
        y='CANT_FACT_ENTREGA_TOTAL',
        data=grouped_data,
        color='orange',
        label='Facturas Entregadas',
        alpha=0.6  # Ajusta la transparencia para ver ambas barras claramente
    )
    
    # Etiquetas al lado de las barras
    for container in ax.containers:
        for bar in container:
            height = bar.get_height()
            ax.text(
                bar.get_x() + bar.get_width(),  # PosiciÃ³n x (justo al lado de la barra)
                bar.get_y() + (height / 1 ),              # PosiciÃ³n y (centro vertical de la barra)
                f'{height}',                           # Etiqueta
                va='center',                           # AlineaciÃ³n vertical
                ha='left',                             # AlineaciÃ³n horizontal
                color='black',                         # Color de la etiqueta
                fontsize=12                            # TamaÃ±o de la fuente
            )
    
    # Etiquetas del grÃ¡fico
    ax.set(xlabel='REPARTIDOR', ylabel='CANTIDAD FACTURAS', title=f'Cantidad de Facturas y Entregadas en el AÃ±o {year}')
    
    # Ajustar el diseÃ±o
    plt.tight_layout()
    
    # Mostrar el grÃ¡fico
    plt.legend()
    plt.show()
```

```python
for year in years:
    plt.figure(figsize=(12,6))
    
    # Supongamos que tienes un DataFrame llamado df con las columnas 'codigo_repartidor', 'facturas_entregadas', 'facturas_totales'
    df_filtered = df[df['YEAR'] == year]
    
    # Agrupar los datos por 'codigo_repartidor' y calcular la suma de 'facturas_entregadas' y 'facturas_totales'
    facturas_por_repartidor = df_filtered.groupby('ID_REPARTIDOR')[['CANT_FACT_ENTREGA_TOTAL', 'CANT_FACT_TOTALES', 'CANT_FACT_ENTREGA_PARCIAL','CANT_FACT_ENTREGA_MOTIVOS']].sum().reset_index()

    # Convertir los datos al formato largo para que seaborn pueda entenderlos mejor
    facturas_long = pd.melt(facturas_por_repartidor, id_vars=['ID_REPARTIDOR'], value_vars=['CANT_FACT_ENTREGA_TOTAL', 'CANT_FACT_TOTALES', 'CANT_FACT_ENTREGA_PARCIAL','CANT_FACT_ENTREGA_MOTIVOS'],
                            var_name='cantidad', value_name='tipo_factura')

    # Crear el grÃ¡fico de barras
    plt.figure(figsize=(12, 6))
    ax = sns.barplot(x='ID_REPARTIDOR', y='tipo_factura', hue='cantidad', data=facturas_long)

    # Personalizar el grÃ¡fico
    ax.set(xlabel='CÃ³digo de Repartidor', ylabel='Cantidad', title=f'Cantidad de Facturas y Entregadas en el AÃ±o {year}')
    

    # AÃ±adir etiquetas a las barras
    for container in ax.containers:
        ax.bar_label(container, size=10)

    # Ajustar el layout
    plt.tight_layout()
    plt.xticks(rotation=45)

    # Mostrar el grÃ¡fico
    plt.show()
```

```python
df_corr = df.select_dtypes(include=['int64','float64'])

correlation_matrix = df_corr.corr()
# Establecer un umbral para considerar una correlaciÃ³n como fuerte
plt.figure(figsize=(20,6))
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm',vmin=-1,vmax=1)
plt.title('Mapa de calor de correlacion')
plt.show()
threshold = 0.6

# Encontrar pares de columnas con correlaciÃ³n fuerte
strong_correlations = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) > threshold:
            colname1 = correlation_matrix.columns[i]
            colname2 = correlation_matrix.columns[j]
            correlation_value = correlation_matrix.iloc[i, j]
            strong_correlations.append((colname1, colname2, correlation_value))

# Imprimir los resultados
print("Pares de columnas con correlaciÃ³n fuerte:")
for col1, col2, corr in strong_correlations:
    print(f"{col1} y {col2}: {corr:.2f}")
```
